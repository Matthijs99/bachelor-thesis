```python
%matplotlib notebook
import math
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
for file_name in ["06445.txt", "06440.txt", "06450.txt"]:
    f  = open(file_name, "r") 
    values = []
    for line in f:
        values.append(int(line))
    n = []
    Fn = []
    for i in range(130, 470):
        n.append(i/100)
        Fn_value = len([value for value in values if value > 10**(i/100)])/len(values)
        Fn.append(math.log10(Fn_value))
    plt.plot(n, Fn, label="p= " + str(int(file_name[:-4])/10000))
    
plt.xlabel("log n")
plt.ylabel("log F(n)")
plt.title("Simulatie voor standaard georiÃ«nteerde percolatie")
plt.legend()
plt.show()
```

```python
def poly(x, a, b):
    return a*x+b
```

```python
%matplotlib notebook
import math
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
#for file_name in ["0735.txt", "074.txt", "0745.txt", "0741.txt", "0742.txt", "0739.txt", "07405.txt", "07395.txt"]:
color_scheme = ["red", "green", "blue", "orange"]
c = 0
for file_name in ["07395.txt", "07400.txt", "07405.txt", "0741.txt"]:
    f  = open(file_name, "r") 
    values = []
    for line in f:
        values.append(int(line))
    n = []
    Fn = []
    for i in range(1300, 4699):
        n.append(i/1000)
        Fn_value = len([value for value in values if value > 10**(i/1000)])/len(values)
        Fn.append(math.log10(Fn_value))
    plt.plot(n, Fn, c=color_scheme[c], label="p= " + str(int(file_name[:-4])/10000))
    ab = np.polyfit(n, Fn, 1, full=True)[0]
    a = ab[0]
    b = ab[1]
    plt.plot(np.arange(1.3,4.68, 0.01), [poly(x, a, b) for x in np.arange(1.3,4.68, 0.01)], '--', c=color_scheme[c])
    c+=1
plt.xlabel("log n")
plt.ylabel("log F(n)")
plt.title("Simulatie voor ons georiÃ«nteerde rooster met eerstegraads polynoom fit")
plt.legend()
plt.show()
```
